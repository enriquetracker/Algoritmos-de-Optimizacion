# Medir el tiempo de ejecución

Medir el tiempo de ejecución de un programa es una tarea muy importante, ya que programar no sólo consiste en crear código que funcione, sino código que pueda ser ejecutado en un tiempo razonable.

La librería **time** nos permite medir el tiempo de ejecución en Python, utilizando el método  `time()`.

```python
# Medida del tiempo de ejecucion con time
import time
inicio = time.time()

# Código a medir
time.sleep(10)
# -------------

fin = time.time()
print(fin-inicio) # 1.0005340576171875
```

## Implementa  una solución para el problema del cambio de moneda en un sistema monetario con las siguientes monedas: {1, 2, 5, 10, 20 , 50}

```python
# Problema de cambio de monedas
def cambio_correcto(cambio):
  dic_cambio = {50:0,20:0,10:0,5:0,2:0,1:0}

  for key,value in dic_cambio.items():

    dic_cambio[key] = cambio//key
    cambio = cambio - dic_cambio[key]*key

  return dic_cambio
```

```python
print(cambio_correcto(88))
```

## Opción 1: Implementa una solución al problema de las 8 reinas en un tablero 8x8

```python
def Verificar(tablero, fila, columna):

    # Verificar si no hay otra reina en la misma columna
    for i in range(len(tablero)):
        if tablero[i,columna] == 1:
            return False

    # Verificar la diagonal superior izquierda
    i = fila - 1
    j = columna - 1
    while i >= 0 and j >= 0:
        if tablero[i,j] == 1:
            return False
        i -= 1
        j -= 1

    # Verificar la diagonal superior derecha
    i = fila - 1
    j = columna + 1
    while i >= 0 and j < 8:
        if tablero[i,j] == 1:
            return False
        i -= 1
        j += 1

    return True
```

```python
def sol_8_reinas(tablero,y):
    global tablero_juego

    if y == 8:
        tablero_juego = tablero
        return True

    for x in range(8):
        if Verificar(tablero, y, x):
            # Colocar una reina en la posición actual
            tablero[y,x] = 1
            nueva_y=y+1
            # Resolver el problema para la siguiente fila
            if sol_8_reinas(tablero, nueva_y):
                return True
            tablero[y, x] = 0

    return False
```
