```python
# Problema de cambio de monedas
def cambio_correcto(cambio):
  dic_cambio = {50:0,20:0,10:0,5:0,2:0,1:0}

  for key,value in dic_cambio.items():

    dic_cambio[key] = cambio//key
    cambio = cambio - dic_cambio[key]*key

  return dic_cambio

print(cambio_correcto(88))
```

```python
def Verificar(tablero, fila, columna):

    # Verificar si no hay otra reina en la misma columna
    for i in range(len(tablero)):
        if tablero[i,columna] == 1:
            return False

    # Verificar la diagonal superior izquierda
    i = fila - 1
    j = columna - 1
    while i >= 0 and j >= 0:
        if tablero[i,j] == 1:
            return False
        i -= 1
        j -= 1

    # Verificar la diagonal superior derecha
    i = fila - 1
    j = columna + 1
    while i >= 0 and j < 8:
        if tablero[i,j] == 1:
            return False
        i -= 1
        j += 1

    return True
```

```python
def sol_8_reinas(tablero,y):
    global tablero_juego

    if y == 8:
        tablero_juego = tablero
        return True

    for x in range(8):
        if Verificar(tablero, y, x):
            # Colocar una reina en la posición actual
            tablero[y,x] = 1
            nueva_y=y+1
            # Resolver el problema para la siguiente fila
            if sol_8_reinas(tablero, nueva_y):
                return True
            tablero[y, x] = 0

    return False
```

# Reto VC-4

Implemente el código del descenso por gradiente para la regresión lineal.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(0,1,20)

def f(x): return x*2

noise_variance =0.2
noise = np.random.randn(x.shape[0])*noise_variance
y = f(x) + noise
```

```python
def coste(x, y, a, b):
    m = len(x)
    error = 0.0
    for i in range(m):
        hipotesis = a+b*x[i]
        error +=  (y[i] - hipotesis) ** 2
    return error / (2*m)
```

```python
def descenso_gradiente(x, y, a, b, alpha, epochs):
    m = len(x)
    hist_coste = []
    for ep in range(epochs):
        b_deriv = 0
        a_deriv = 0
        for i in range(m):
            hipotesis = a+b*x[i]
            a_deriv += hipotesis - y[i]
            b_deriv += (hipotesis - y[i]) * x[i]
            hist_coste.append(coste(x, y, a, b))
        a -= (a_deriv / m) * alpha
        b -= (b_deriv / m) * alpha

    return a, b, hist_coste
```

```python
a=1
b=1
alpha = 0.0001
iters = 50000
a,b, hist_coste = descenso_gradiente(x, y, a, b, alpha, iters)
```

```python
plt.plot(np.arange(0,iters*len(x)),hist_coste, 'r-')
```

```python
import math
import random
#Definimos la funcion
f= lambda X: math.sin(1/2 * X[0]**2 - 1/4 * X[1]**2 + 3) *math.cos(2*X[0] + 1 - math.exp(X[1]) )
df = lambda X: [X[0]*math.cos(2*X[0] - math.exp(X[1])+1)*math.cos(X[0]**2/2 - X[1]**2/4 + 3) - 2*math.sin(2*X[0] - math.exp(X[1]) +1)*math.sin(X[0]**2/2 - X[1]**2/4 + 3) ,math.exp(X[1])*math.sin(-X[0]**2/2 + X[1]**2/4 - 3)*math.cos(-2*X[0] + math.exp(X[1])-1)- (X[1]*math.cos(- X[0]**2/2 + X[1]**2/4 - 3)*math.cos(-2*X[0] + math.exp(X[1])-1))/2]
```

```python
#Prepara los datos para dibujar mapa de niveles de Z
resolucion = 100
rango=5.5

X=np.linspace(-rango,rango,resolucion)
Y=np.linspace(-rango,rango,resolucion)
Z=np.zeros((resolucion,resolucion))
for ix,x in enumerate(X):
  for iy,y in enumerate(Y):
    Z[iy,ix] = f([x,y])

#Pinta el mapa de niveles de Z
plt.contourf(X,Y,Z,resolucion)
plt.colorbar()

#Generamos un punto aleatorio inicial y pintamos de blanco
P=[random.uniform(-5,5  ),random.uniform(-5,5 ) ]
plt.plot(P[0],P[1],"o",c="white")

#Tasa de aprendizaje. Fija. Sería más efectivo reducirlo a medida que nos acercamos.
TA=0.1

#Iteraciones:50
for _ in range(50):
  grad = df(P)
  #print(P,grad)
  P[0],P[1] = P[0] - TA*grad[0] , P[1] - TA*grad[1]
  plt.plot(P[0],P[1],"o",c="red")

#Dibujamos el punto final y pintamos de verde
plt.plot(P[0],P[1],"o",c="green")
plt.show()
print("Solucion:" , P , f(P))
```
